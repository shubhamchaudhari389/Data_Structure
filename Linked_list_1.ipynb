### Write A Python Program To Create Single Linked List And Returning That Single Linked List ?

```python
# Define a Node class to represent elements in the linked list
class Node:
    def __init__(self, data=None):
        self.data = data  # Data of the node
        self.next = None   # Reference to the next node in the linked list

# Define a LinkedList class to manage the linked list
class LinkedList:
    def __init__(self):
        self.head = None  # Initialize an empty linked list with no head

    # Method to append a new node with given data to the end of the linked list
    def append(self, data):
        new_node = Node(data)
        if not self.head:  # If the linked list is empty, set the new node as the head
            self.head = new_node
            return
        last_node = self.head
        while last_node.next:  # Iterate to the last node in the linked list
            last_node = last_node.next
        last_node.next = new_node  # Add the new node to the end of the linked list

    # Method to display the elements of the linked list
    def display(self):
        current = self.head
        while current:  # Iterate through the linked list
            print(current.data, end=" -> ")  # Print the data of the current node
            current = current.next  # Move to the next node
        print("None")  # Print "None" to indicate the end of the linked list

    # Method to get the head of the linked list
    def get_linked_list(self):
        return self.head

# Example usage of the LinkedList class
if __name__ == "__main__":
    my_linked_list = LinkedList()  # Create an instance of the LinkedList class

    # Append elements to the linked list
    my_linked_list.append(10)
    my_linked_list.append(20)
    my_linked_list.append(30)
    my_linked_list.append(40)
    my_linked_list.append(50)
    my_linked_list.append(60)

    print("Linked List:")
    my_linked_list.display()  # Display the elements of the linked list

    returned_linked_list = my_linked_list.get_linked_list()  # Get the head of the linked list
    print("\nReturned Linked List:")
    while returned_linked_list:  # Iterate through the returned linked list
        print(returned_linked_list.data, end=" -> ")  # Print the data of the current node
        returned_linked_list = returned_linked_list.next  # Move to the next node
    print("None")  # Print "None" to indicate the end of the linked list
```

### Write A Python Program To Find Address Of A Node Which Contain Data _ X In The Given Single Linked List(1st -X Only if N0 Of X Then Returing Null )

### Write A Program To Find address Of Last 2nd last Node In The Given Sinlge Linked List

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    def get_last_two_node_addresses(self):
        if not self.head or not self.head.next:
            return None, None  # Return None for both addresses if the list is empty or has only one node

        last_node = self.head
        second_last_node = None

        while last_node.next:
            second_last_node = last_node
            last_node = last_node.next

        return id(second_last_node), id(last_node)

# Example usage:
if __name__ == "__main__":
    my_linked_list = LinkedList()
    my_linked_list.append(1)
    my_linked_list.append(2)
    my_linked_list.append(3)
    my_linked_list.append(4)

    print("Linked List:")
    my_linked_list.display()

    second_last_address, last_address = my_linked_list.get_last_two_node_addresses()

    if second_last_address is not None:
        print(f"\nAddress of Second-to-Last Node: {second_last_address}")
    else:
        print("\nLinked list has less than two nodes.")

    if last_address is not None:
        print(f"Address of Last Node: {last_address}")
    else:
        print("Linked list is empty.")
```

### Write A Python Program To Insert A Node With Data-X Before A ode With Data-Y In The Given Single Linked List (Before 1st - Y If Node Y Then No-X)

```python
# Definition of the Node class representing individual nodes in the linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Definition of the LinkedList class representing the linked list
class LinkedList:
    def __init__(self):
        self.head = None  # Initialization of the linked list with an empty head

    def display(self):
        # Display method to print the elements of the linked list
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

    def insert_before(self, x, y):
        # Method to insert a node with data x before a node with data y in the linked list
        new_node = Node(x)  # Create a new node with data x

        # If the list is empty or Y is the first node
        if not self.head or self.head.data == y:
            new_node.next = self.head
            self.head = new_node
            return

        current = self.head
        while current.next:
            # Find the node with data y
            if current.next.data == y:
                new_node.next = current.next
                current.next = new_node
                return
            current = current.next

        # If Y is not found, insert at the end
        current.next = new_node

# Example usage:
linked_list = LinkedList()  # Create an instance of the LinkedList class

# Insert nodes with data 2, 4, 6, and 8 before nodes with data 3 and 4
linked_list.insert_before(2, 3)
linked_list.insert_before(4, 3)
linked_list.insert_before(6, 3)
linked_list.insert_before(8, 4)

print("Original Linked List:")
linked_list.display()  # Display the original linked list

# Insert a node with data 5 before the first occurrence of a node with data 4
linked_list.insert_before(5, 4)

print("\nLinked List after insertion:")
linked_list.display()  # Display the linked list after the insertion
```
