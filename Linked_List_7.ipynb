### Write A Python Program To Insert A Node wITH data X at The End Of C-SLL ?

```python
# Define a Node class to represent individual nodes in the linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Define a CircularSinglyLinkedList class to manage the circular linked list
class CircularSinglyLinkedList:
    def __init__(self):
        # Initialize an empty circular linked list with a head set to None
        self.head = None

    def insert_at_end(self, x):
        # Create a new node with data x
        new_node = Node(x)

        # Check if the list is empty
        if not self.head:
            # If empty, set the new node as the head and create a loop by pointing its next to itself
            self.head = new_node
            self.head.next = self.head
        else:
            # If not empty, traverse the list to find the last node
            temp = self.head
            while temp.next != self.head:
                temp = temp.next

            # Update the last node's next to point to the new node
            temp.next = new_node
            # Complete the circular structure by pointing the new node's next to the head
            new_node.next = self.head

    def display_list(self):
        # Display the elements of the circular linked list
        if not self.head:
            print("List is empty")
            return

        temp = self.head
        while True:
            # Print the data of the current node
            print(temp.data, end=' ')
            # Move to the next node
            temp = temp.next
            # Check if we have reached the head again, indicating the end of the circular list
            if temp == self.head:
                break
        print()

# Example usage:
if __name__ == "__main__":
    # Create an instance of the CircularSinglyLinkedList class
    circular_list = CircularSinglyLinkedList()

    # Insert nodes with data 10, 20, and 30 at the end of the circular linked list
    circular_list.insert_at_end(10)
    circular_list.insert_at_end(20)
    circular_list.insert_at_end(30)

    # Display the circular linked list after insertion
    print("Circular Linked List after insertion:")
    circular_list.display_list()
```

### Write A python Program To Delete First Node Of Circular Linked List 

```python
# Define a Node class to represent individual nodes in the linked list
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Define a CircularSinglyLinkedList class to manage the circular linked list
class CircularSinglyLinkedList:
    def __init__(self):
        # Initialize an empty circular linked list with a head set to None
        self.head = None

    def insert_at_end(self, x):
        # Create a new node with data x
        new_node = Node(x)

        # Check if the list is empty
        if not self.head:
            # If empty, set the new node as the head and create a loop by pointing its next to itself
            self.head = new_node
            self.head.next = self.head
        else:
            # If not empty, traverse the list to find the last node
            temp = self.head
            while temp.next != self.head:
                temp = temp.next

            # Update the last node's next to point to the new node
            temp.next = new_node
            # Complete the circular structure by pointing the new node's next to the head
            new_node.next = self.head

    def delete_first_node(self):
        # Check if the list is empty
        if not self.head:
            print("List is empty. Nothing to delete.")
            return

        # If the list has only one node, set head to None
        if self.head.next == self.head:
            self.head = None
        else:
            # Find the last node in the circular list
            temp = self.head
            while temp.next != self.head:
                temp = temp.next

            # Move head to the next node and update the last node's next to the new head
            self.head = self.head.next
            temp.next = self.head

    def display_list(self):
        # Display the elements of the circular linked list
        if not self.head:
            print("List is empty")
            return

        temp = self.head
        while True:
            # Print the data of the current node
            print(temp.data, end=' ')
            # Move to the next node
            temp = temp.next
            # Check if we have reached the head again, indicating the end of the circular list
            if temp == self.head:
                break
        print()

# Example usage:
if __name__ == "__main__":
    # Create an instance of the CircularSinglyLinkedList class
    circular_list = CircularSinglyLinkedList()

    # Insert nodes with data 10, 20, and 30 at the end of the circular linked list
    circular_list.insert_at_end(10)
    circular_list.insert_at_end(20)
    circular_list.insert_at_end(30)

    # Display the circular linked list before deletion
    print("Circular Linked List before deletion:")
    circular_list.display_list()

    # Delete the first node
    circular_list.delete_first_node()

    # Display the circular linked list after deletion
    print("Circular Linked List after deletion:")
    circular_list.display_list()
```
