Write A python Program To Delete A node Which Contain Data-X in the given Single Linked List

```python
# Define a Node class to represent individual nodes in the linked list
class Node:
    def __init__(self, data):
        self.data = data  # Initialize node data
        self.next = None   # Initialize the reference to the next node as None

# Define a LinkedList class to represent the linked list
class LinkedList:
    def __init__(self):
        self.head = None  # Initialize the head of the linked list as None

    def append(self, data):
        # Method to add a new node with given data at the end of the linked list
        new_node = Node(data)  # Create a new node with the provided data
        if not self.head:
            self.head = new_node  # If the list is empty, set the new node as the head
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node  # Attach the new node to the end of the list

    def delete_node(self, key):
        # Method to delete a node with the given data from the linked list
        current_node = self.head

        # If the node to be deleted is the head
        if current_node and current_node.data == key:
            self.head = current_node.next  # Update the head to the next node
            current_node = None  # Remove the reference to the deleted node
            return

        # Search for the node to be deleted
        prev_node = None
        while current_node and current_node.data != key:
            prev_node = current_node
            current_node = current_node.next

        # If the node with the given data is not found
        if current_node is None:
            print("Node with data {} not found in the list.".format(key))
            return

        # Unlink the node from the linked list
        prev_node.next = current_node.next
        current_node = None  # Remove the reference to the deleted node

    def display_list(self):
        # Method to display the elements of the linked list
        current_node = self.head
        while current_node:
            print(current_node.data, end=" ")  # Print the data of the current node
            current_node = current_node.next
        print()

# Example Usage:
linked_list = LinkedList()  # Create an instance of the LinkedList class
linked_list.append(1)
linked_list.append(2)
linked_list.append(3)
linked_list.append(4)

print("Original Linked List:")
linked_list.display_list()  # Display the original linked list

# Delete node with data 2
linked_list.delete_node(2)  # Delete the node with data 2

print("\nLinked List after deleting node with data 2:")
linked_list.display_list()  # Display the linked list after deletion
```

Write A Python Program To Move Last Node Of The Single Linked List To The FRont Off The Single Linked List

```python
# Define a Node class to represent individual nodes in the linked list
class Node:
    def __init__(self, data):
        self.data = data        # Initialize node data
        self.next = None         # Initialize the reference to the next node as None

# Define a LinkedList class to represent the linked list
class LinkedList:
    def __init__(self):
        self.head = None         # Initialize the head of the linked list as None

    def append(self, data):
        new_node = Node(data)    # Create a new node with the provided data
        if not self.head:
            self.head = new_node  # If the list is empty, set the new node as the head
            return
        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node  # Attach the new node to the end of the list

    def move_last_to_front(self):
        if not self.head or not self.head.next:
            # If the list is empty or has only one element, no need to move.
            return

        last_node = None
        second_last = None
        current_node = self.head

        # Traverse the list to find the last and second-to-last nodes
        while current_node.next:
            second_last = current_node
            current_node = current_node.next

        last_node = current_node
        second_last.next = None  # Disconnect the last node

        # Move the last node to the front
        last_node.next = self.head
        self.head = last_node

    def display_list(self):
        current_node = self.head
        while current_node:
            print(current_node.data, end=" ")  # Print the data of the current node
            current_node = current_node.next
        print()


# Example Usage:
linked_list = LinkedList()
linked_list.append(1)
linked_list.append(2)
linked_list.append(3)
linked_list.append(4)

print("Original Linked List:")
linked_list.display_list()

# Move the last node to the front
linked_list.move_last_to_front()

print("\nLinked List after moving the last node to the front:")
linked_list.display_list()
```
