### Write A Python Program To Insert A Node With data-X before A Node With Data-y In Doubal Linked LIst?

```python
# Define a Node class to represent nodes in a doubly linked list
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
        self.prev = None

# Define a DoublyLinkedList class
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Method to insert a node with data X before a node with data Y
    def insert_before(self, x, y):
        # Create a new node with data X
        new_node = Node(x)

        # Check if the list is empty
        if not self.head:
            print(f"Error: Node with data {y} not found in the list.")
            return

        # Traverse the list to find the node with data Y
        current = self.head
        while current:
            if current.data == y:
                # Insert the new node before the node with data Y
                if current.prev:
                    current.prev.next = new_node
                    new_node.prev = current.prev
                else:
                    # If Y is the head, update the head to be the new node
                    self.head = new_node

                new_node.next = current
                current.prev = new_node

                print(f"Node with data {x} inserted before the node with data {y}.")
                return

            current = current.next

        # If the node with data Y is not found
        print(f"Error: Node with data {y} not found in the list.")

    # Method to display the doubly linked list
    def display(self):
        current = self.head
        while current:
            print(current.data, end=" ")
            current = current.next
        print()

# Example Usage
dll = DoublyLinkedList()

# Create nodes and link them to form a doubly linked list
dll.head = Node(1)
second_node = Node(2)
third_node = Node(4)

dll.head.next = second_node
second_node.prev = dll.head
second_node.next = third_node
third_node.prev = second_node

print("Original Doubly Linked List:")
dll.display()

# Define values for insertion
x_value = 3
y_value = 4

# Call the insert_before method to insert a node with data X before a node with data Y
dll.insert_before(x_value, y_value)

print("Doubly Linked List after insertion:")
dll.display()
```

### Write A PYthon PRogram To Delete 1st Node Of Circular Doubal Linked List ?

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
        self.prev = None

class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            new_node.prev = self.head
        else:
            last = self.head.prev
            last.next = new_node
            new_node.prev = last
            new_node.next = self.head
            self.head.prev = new_node

    def delete_first(self):
        if not self.head:
            print("Error: Circular Doubly Linked List is empty.")
            return

        if self.head.next == self.head:  # Only one node in the list
            self.head = None
        else:
            first = self.head
            second = self.head.next

            self.head = second
            second.prev = first.prev
            first.prev.next = second

        print("First node deleted.")

    def display(self):
        if not self.head:
            print("Circular Doubly Linked List is empty.")
            return

        current = self.head
        while True:
            print(current.data, end=" ")
            current = current.next
            if current == self.head:
                break
        print()

# Example Usage
cdll = CircularDoublyLinkedList()

# Append nodes to create a circular doubly linked list
cdll.append(1)
cdll.append(2)
cdll.append(3)

print("Original Circular Doubly Linked List:")
cdll.display()

# Delete the first node
cdll.delete_first()

print("Circular Doubly Linked List after deletion:")
cdll.display()
```

### Write A Python Program To Delete 1st Node Of Circular Linked List.

```python
# Define a Node class to represent nodes in a circular doubly linked list
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
        self.prev = None

# Define a CircularDoublyLinkedList class
class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None

    # Method to append a node to the circular doubly linked list
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            # If the list is empty, set the new node as the head and make it circular
            self.head = new_node
            new_node.next = self.head
            new_node.prev = self.head
        else:
            # If the list is not empty, insert the new node at the end and update links
            last = self.head.prev
            last.next = new_node
            new_node.prev = last
            new_node.next = self.head
            self.head.prev = new_node

    # Method to delete the first node of the circular doubly linked list
    def delete_first(self):
        if not self.head:
            print("Error: Circular Doubly Linked List is empty.")
            return

        if self.head.next == self.head:  # Only one node in the list
            self.head = None
        else:
            # If there are more than one node, update links to delete the first node
            first = self.head
            second = self.head.next

            self.head = second
            second.prev = first.prev
            first.prev.next = second

        print("First node deleted.")

    # Method to display the circular doubly linked list
    def display(self):
        if not self.head:
            print("Circular Doubly Linked List is empty.")
            return

        current = self.head
        while True:
            print(current.data, end=" ")
            current = current.next
            if current == self.head:
                break
        print()

# Example Usage
cdll = CircularDoublyLinkedList()

# Append nodes to create a circular doubly linked list
cdll.append(1)
cdll.append(2)
cdll.append(3)

print("Original Circular Doubly Linked List:")
cdll.display()

# Delete the first node
cdll.delete_first()

print("Circular Doubly Linked List after deletion:")
cdll.display()
```
